```python
path = "/content/drive/My Drive/Pilot-attendant/label-extraction/fa_male/Flight_attendant_Male_Light"
```

```python
from google.oauth2 import service_account

credentials = service_account.Credentials.from_service_account_file(
'/content/drive/My Drive/Pilot-attendant/label-extraction/service-account.json')

files = os.listdir(path)

label_fa_d = open("/content/drive/My Drive/Pilot-attendant/label-extraction/fa_male/mfa_Light.txt", "w+")

for f in files:

  client = vision.ImageAnnotatorClient(credentials = credentials)

  image_path = f'{path}/{f}'

  with io.open(image_path, 'rb') as image_file:
    content = image_file.read()

  image = vision.Image(content = content) 

  response = client.label_detection(image=image)
  label_fa_d.write('mfa ' + '.' + f + '\n')
  label_fa_d.write(' \t' + 'Label' + ' \t ' + 'Score' + '\n')

  i = 0

  for label in response.label_annotations:
      line = str(i) + '\t' + f'{label.description} \t {(label.score*100.)} \n' 
      label_fa_d.write(line)
      i = i + 1

  resp = client.object_localization(image=image)
  label_fa_d.write('\t' + 'Object' + ' \t ' + 'Score' + '\n')

  j = 0

  for obj in resp.localized_object_annotations:
      line = str(j) + '\t' + f'{obj.name} \t {(obj.score*100.)} \n' 
      label_fa_d.write(line)
      j = j + 1
  label_fa_d.write('\n\n')


label_fa_d.close()
```
